```python
# ============================================================
# CREACIÓN DE TARGETS PARA MODELADO
# ============================================================
# Creamos dos posibles variables objetivo:
#
# 1. target_binary:
#    - 1 = el oro sube al día siguiente
#    - 0 = el oro no sube al día siguiente
#
# 2. target_multiclass:
#    - 1  = comprar / sube de forma significativa
#    - 0  = mantener / movimiento pequeño
#    - -1 = vender / baja de forma significativa
#
# Así dejamos preparado el dataset para que modelado pueda probar
# ambos enfoques y decidir cuál funciona mejor.
# ============================================================

# Calculamos la rentabilidad del oro al día siguiente.
# Esta será la base para construir ambos targets.
df["gold_return_next_day"] = (
    (df["gold_close"].shift(-1) - df["gold_close"]) / df["gold_close"]
)

# ------------------------------------------------------------
# TARGET BINARIO
# ------------------------------------------------------------
# Si el retorno de mañana es mayor que 0, el oro sube.
# Si es menor o igual que 0, no sube.
df["target_binary"] = np.where(
    df["gold_return_next_day"] > 0,
    1,
    0
)

# ------------------------------------------------------------
# TARGET MULTICLASE
# ------------------------------------------------------------
# Definimos un umbral para separar movimientos pequeños
# de movimientos más relevantes.
#
# 0.002 equivale a 0.2%.
umbral = 0.002

# Creamos el target multiclase:
#  1  = comprar
#  0  = mantener
# -1  = vender
df["target_multiclass"] = np.where(
    df["gold_return_next_day"] > umbral,
    1,
    np.where(
        df["gold_return_next_day"] < -umbral,
        -1,
        0
    )
)

# Comprobamos las primeras filas
df[[
    "Date",
    "gold_close",
    "gold_return_next_day",
    "target_binary",
    "target_multiclass"
]].head(10)
```

```python
# ============================================================
# DISTRIBUCIÓN DE LOS TARGETS
# ============================================================
# Revisamos cuántos registros hay en cada clase.
#
# Esto es importante porque si una clase tiene muy pocos ejemplos,
# el modelo puede aprender peor esa categoría.
# ============================================================

print("Distribución del target binario:")
print(df["target_binary"].value_counts())

print("\nDistribución porcentual del target binario:")
print(df["target_binary"].value_counts(normalize=True) * 100)

print("\nDistribución del target multiclase:")
print(df["target_multiclass"].value_counts())

print("\nDistribución porcentual del target multiclase:")
print(df["target_multiclass"].value_counts(normalize=True) * 100)
```

- 52,9% de los casos corresponden a subidas del precio del oro.
- 47,1% corresponden a no subida.

**Target multiclase**
- 42,1% Comprar.

Para construir el target multiclase se estableció inicialmente un umbral del 0,2% para diferenciar movimientos relevantes de pequeñas variaciones diarias.

Tras analizar la distribución de las clases obtenidas, se observó que las tres categorías mantenían una representación suficiente y relativamente equilibrada, por lo que se decidió conservar dicho umbral para el resto del proyecto.

Las distribuciones obtenidas permiten entrenar modelos de clasificación sin necesidad de aplicar técnicas adicionales de balanceo en esta fase del proyecto.

## Feature Engineering

En esta sección creamos nuevas variables a partir de las columnas originales.

El objetivo es transformar los datos de mercado en señales más útiles para el modelo, evitando usar información futura.

Las nuevas variables se calculan siempre con información disponible en el mismo día o en días anteriores.

```python
# ============================================================
# FEATURE ENGINEERING: RETORNOS DIARIOS
# ============================================================
# Calculamos el retorno diario de cada activo/indicador.
#
# El retorno diario mide cuánto cambia una variable respecto
# al día anterior.
#
# Fórmula:
# retorno = (precio_hoy - precio_ayer) / precio_ayer
#
# Esto suele ser más útil que usar solo precios absolutos,
# porque permite al modelo aprender movimientos relativos.
# ============================================================

# Retorno diario del oro
df["gold_return"] = df["gold_close"].pct_change()

# Retorno diario del índice dólar DXY
df["dxy_return"] = df["dxy_close"].pct_change()

# Retorno diario del VIX
df["vix_return"] = df["vix_close"].pct_change()

# Retorno diario del TNX
df["tnx_return"] = df["tnx_close"].pct_change()

# Mostramos las primeras filas para comprobar el resultado
df[[
    "Date",
    "gold_close",
    "gold_return",
    "dxy_close",
    "dxy_return",
    "vix_close",
    "vix_return",
    "tnx_close",
    "tnx_return"
]].head(10)
```

```python
# ============================================================
# FEATURE ENGINEERING: RANGO DIARIO
# ============================================================
# Calculamos el rango diario de cada activo/indicador.
#
# El rango diario mide la diferencia entre el precio máximo
# y el precio mínimo del mismo día.
#
# Fórmula:
# rango = máximo del día - mínimo del día
#
# Esta variable puede ayudar al modelo a captar volatilidad
# o movimientos fuertes dentro de una jornada.
# ============================================================

# Rango diario del oro
df["gold_range"] = df["gold_high"] - df["gold_low"]

# Rango diario del índice dólar DXY
df["dxy_range"] = df["dxy_high"] - df["dxy_low"]

# Rango diario del VIX
df["vix_range"] = df["vix_high"] - df["vix_low"]

# Rango diario del TNX
df["tnx_range"] = df["tnx_high"] - df["tnx_low"]

# Comprobamos las nuevas variables creadas
df[[
    "Date",
    "gold_range",
    "dxy_range",
    "vix_range",
    "tnx_range"
]].head(10)
```

```python
# ============================================================
# FEATURE ENGINEERING: VARIACIÓN APERTURA-CIERRE
# ============================================================
# Calculamos cuánto se mueve cada activo dentro del mismo día,
# comparando el cierre con la apertura.
#
# Fórmula:
# variación apertura-cierre = (cierre - apertura) / apertura
#
# Esta variable nos dice si durante la jornada el activo terminó
# subiendo o bajando respecto a cómo empezó.
# ============================================================

# Variación apertura-cierre del oro
df["gold_open_close_return"] = (
    (df["gold_close"] - df["gold_open"]) / df["gold_open"]
)

# Variación apertura-cierre del índice dólar DXY
df["dxy_open_close_return"] = (
    (df["dxy_close"] - df["dxy_open"]) / df["dxy_open"]
)

# Variación apertura-cierre del VIX
df["vix_open_close_return"] = (
    (df["vix_close"] - df["vix_open"]) / df["vix_open"]
)

# Variación apertura-cierre del TNX
df["tnx_open_close_return"] = (
    (df["tnx_close"] - df["tnx_open"]) / df["tnx_open"]
)

# Comprobamos las nuevas variables creadas
df[[
    "Date",
    "gold_open_close_return",
    "dxy_open_close_return",
    "vix_open_close_return",
    "tnx_open_close_return"
]].head(10)
```

```python
# ============================================================
# FEATURE ENGINEERING: MEDIAS MÓVILES DEL ORO
# ============================================================
# Calculamos medias móviles del precio de cierre del oro.
#
# Una media móvil suaviza el precio y permite ver la tendencia
# de los últimos días.
#
# Usamos ventanas de:
# - 5 días: tendencia corta
# - 20 días: tendencia más amplia
#
# Estas variables solo usan información del día actual y días
# anteriores, por lo que no introducen data leakage.
# ============================================================

# Media móvil de 5 días del oro
df["gold_ma_5"] = df["gold_close"].rolling(window=5).mean()

# Media móvil de 20 días del oro
df["gold_ma_20"] = df["gold_close"].rolling(window=20).mean()

# Diferencia relativa entre el precio actual y la media móvil de 5 días
df["gold_close_vs_ma_5"] = (
    (df["gold_close"] - df["gold_ma_5"]) / df["gold_ma_5"]
)

# Diferencia relativa entre el precio actual y la media móvil de 20 días
df["gold_close_vs_ma_20"] = (
    (df["gold_close"] - df["gold_ma_20"]) / df["gold_ma_20"]
)

# Comprobamos las nuevas variables
df[[
    "Date",
    "gold_close",
    "gold_ma_5",
    "gold_ma_20",
    "gold_close_vs_ma_5",
    "gold_close_vs_ma_20"
]].head(25)
```

### Resumen de variables derivadas

Se han creado nuevas variables a partir de los datos originales con el objetivo de capturar:

- Tendencias de mercado (medias móviles).
- Movimientos relativos (retornos).
- Volatilidad intradía (rangos).
- Comportamiento diario (variación apertura-cierre).

Estas variables serán utilizadas posteriormente durante la fase de modelado para evaluar su capacidad predictiva sobre la evolución futura del precio del oro.

```python
# ============================================================
# REVISIÓN DE NULOS TRAS FEATURE ENGINEERING
# ============================================================
# Algunas features generan nulos de forma natural:
#
# - pct_change() genera nulo en la primera fila porque no hay día anterior.
# - rolling(5) genera nulos en las primeras 4 filas.
# - rolling(20) genera nulos en las primeras 19 filas.
# - gold_return_next_day genera nulo en la última fila porque no hay día siguiente.
#
# Revisamos cuántos nulos hay antes de limpiarlos.
# ============================================================

nulos_post_features = df.isnull().sum()

print("Columnas con nulos después del Feature Engineering:")
print(nulos_post_features[nulos_post_features > 0])

print("\nFilas con al menos un valor nulo:")
print(df.isnull().any(axis=1).sum())
```

### Tratamiento de nulos generados por Feature Engineering

La creación de retornos y medias móviles genera valores nulos de forma natural en los extremos de la serie temporal:

- Los retornos requieren información del día anterior.
- Las medias móviles requieren suficientes observaciones previas.
- El cálculo del retorno futuro necesita información del día siguiente.

Dado que estos nulos representan menos del 1% del conjunto de datos y son consecuencia directa de los cálculos realizados, se opta por eliminar las filas afectadas.

```python
# ============================================================
# ELIMINACIÓN DE NULOS GENERADOS POR FEATURE ENGINEERING
# ============================================================
# Eliminamos las filas afectadas por:
# - retornos diarios
# - medias móviles
# - cálculo del target futuro
#
# Estos nulos son esperados y aparecen únicamente en los
# extremos de la serie temporal.
# ============================================================

filas_antes = len(df)

# Eliminamos filas con nulos
df = df.dropna()

filas_despues = len(df)

print("Filas antes:", filas_antes)
print("Filas después:", filas_despues)
print("Filas eliminadas:", filas_antes - filas_despues)

print("\nNulos restantes:")
print(df.isnull().sum().sum())
```

## Validación de Data Leakage

El data leakage ocurre cuando el modelo recibe información que en la vida real no estaría disponible en el momento de hacer la predicción.

En este proyecto debemos evitar especialmente dos errores:

1. Usar variables que miran al futuro, como `gold_return_next_day`.
2. Mezclar fechas antiguas y recientes de forma aleatoria en entrenamiento y prueba.

Por eso, las variables objetivo y las columnas creadas con información futura se excluyen de las variables predictoras.

```python
# ============================================================
# IDENTIFICACIÓN DE COLUMNAS NO PREDICTORAS
# ============================================================
# Definimos qué columnas NO deben entrar como variables X
# en el entrenamiento del modelo.
#
# Motivos:
# - Date se usa para ordenar temporalmente, no como variable numérica.
# - gold_return_next_day contiene información del futuro.
# - target_binary y target_multiclass son variables objetivo.
# ============================================================

columnas_no_predictoras = [
    "Date",
    "gold_return_next_day",
    "target_binary",
    "target_multiclass"
]

print("Columnas excluidas como predictoras:")
for columna in columnas_no_predictoras:
    print("-", columna)
```

```python
# ============================================================
# CREACIÓN DE MATRICES DE MODELADO
# ============================================================
# Construimos:
#
# X              -> variables predictoras
# y_binary       -> target binario
# y_multiclass   -> target multiclase
#
# Excluimos todas las columnas que contienen información futura
# o que son variables objetivo.
# ============================================================

# Variables predictoras
X = df.drop(columns=columnas_no_predictoras)

# Targets
y_binary = df["target_binary"]

y_multiclass = df["target_multiclass"]

# Comprobaciones
print("Dimensiones de X:", X.shape)
print("Dimensiones de y_binary:", y_binary.shape)
print("Dimensiones de y_multiclass:", y_multiclass.shape)

print("\nPrimeras variables predictoras:")
print(X.columns.tolist())
```

### Resultado de la validación de Data Leakage

Se han separado las variables predictoras de las variables objetivo.

Además, se ha excluido la columna `gold_return_next_day`, ya que contiene información del día siguiente y podría provocar data leakage durante el entrenamiento.

De este modo, las variables utilizadas para el modelado solo contienen información disponible en el momento de realizar la predicción.

## Split temporal Train/Test

Al trabajar con series temporales, la división entre entrenamiento y prueba no debe hacerse de forma aleatoria.

Si mezclamos fechas antiguas y recientes, el modelo podría aprender patrones usando información futura.

Por eso se utiliza un split temporal:

- `train`: datos más antiguos.
- `test`: datos más recientes.

```python
# ============================================================
# TRAIN/TEST SPLIT TEMPORAL
# ============================================================
# Dividimos los datos respetando el orden cronológico.
#
# Usamos el 80% inicial para entrenamiento y el 20% final
# para prueba.
#
# Importante:
# No usamos train_test_split aleatorio porque estamos trabajando
# con series temporales.
# ============================================================

# Calculamos el punto de corte temporal
split_index = int(len(df) * 0.8)

# Dividimos X respetando el orden temporal
X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]

# Dividimos los dos targets con el mismo corte temporal
y_binary_train = y_binary.iloc[:split_index]
y_binary_test = y_binary.iloc[split_index:]

y_multiclass_train = y_multiclass.iloc[:split_index]
y_multiclass_test = y_multiclass.iloc[split_index:]

# También guardamos las fechas para documentar el rango temporal
dates_train = df["Date"].iloc[:split_index]
dates_test = df["Date"].iloc[split_index:]

print("Tamaño X_train:", X_train.shape)
print("Tamaño X_test:", X_test.shape)

print("\nRango temporal train:")
print(dates_train.min().date(), "->", dates_train.max().date())

print("\nRango temporal test:")
print(dates_test.min().date(), "->", dates_test.max().date())
```

### Resultado del split temporal

La división entre entrenamiento y prueba se realizó respetando el orden cronológico de la serie temporal.

- Entrenamiento: 2015-01-30 a 2024-03-12.
- Prueba: 2024-03-13 a 2026-06-25.

De esta forma se evita el uso de información futura durante el entrenamiento y se obtiene una evaluación más realista del rendimiento de los modelos.

## Escalado de variables

Las variables numéricas del dataset presentan escalas muy diferentes.

Por ejemplo, el precio del oro se mueve en miles de unidades, mientras que otros indicadores como TNX o VIX presentan valores mucho menores.

Para evitar que las variables con magnitudes más altas tengan un peso excesivo durante el entrenamiento, se aplica un proceso de estandarización mediante `StandardScaler`.

```python
# ============================================================
# ESCALADO DE VARIABLES
# ============================================================
# Aplicamos StandardScaler utilizando únicamente los datos
# de entrenamiento.
#
# Posteriormente transformamos tanto train como test usando
# los parámetros aprendidos en train.
#
# Esto evita data leakage.
# ============================================================

# Creamos el escalador
scaler = StandardScaler()

# Aprendemos los parámetros con train y escalamos train
X_train_scaled = scaler.fit_transform(X_train)

# Escalamos test con los parametros aprendidos en train
X_test_scaled = scaler.transform(X_test)

print("Shape X_train_scaled:", X_train_scaled.shape)
print("Shape X_test_scaled:", X_test_scaled.shape)
```

## Pipeline de preprocesamiento

Para garantizar la reproducibilidad del proceso se define un pipeline de preprocesamiento.

El pipeline permite aplicar de forma consistente las transformaciones necesarias antes del entrenamiento de los modelos, evitando errores manuales y facilitando la reutilización del flujo de trabajo.

```python
# ============================================================
# IMPORTACIÓN DE PIPELINE
# ============================================================

from sklearn.pipeline import Pipeline
```

```python
# ============================================================
# PIPELINE DE PREPROCESAMIENTO
# ============================================================
# Definimos un pipeline sencillo que aplica el escalado
# mediante StandardScaler.
#
# Este pipeline podrá reutilizarse posteriormente durante
# la fase de modelado.
# ============================================================

preprocessing_pipeline = Pipeline([
    ("scaler", StandardScaler())
])

print(preprocessing_pipeline)
```
